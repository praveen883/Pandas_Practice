```python
new_df = df.fillna("40")
new_df
```

```python
new_df = df.fillna({'temperature':35,
                   'windspeed':15,
                   'event':"Rain"})
print(df)
print(new_df)
```

```python
new_df = df.fillna(method="ffill")
print(df)
print(new_df)
```

```python
print(df)
new_df = df.fillna(method="bfill")
new_df
```

```python
print(df)
new_df = df.fillna(method="ffill", axis="columns") #fill the empty values with the values of previous columns. 
new_df                                             #if we use bfill next column
```

```python
df
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
print(df)
df.interpolate(threshold=1) #interpolate method assigns mean value of previous and next rows to the NaN row.
```

```python
print(df.interpolate())
new_df = df.interpolate(method="time") #fill the NAN values with mean values based on time column.
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how="all") #drop the row only if all columns has NaN
new_df
```

```python
new_df = df.dropna(thresh=2) #thresh argument used drop rows based on no.of fields which has null in it.
print(df)
new_df
```

```python
new_df = df.fillna(0,limit=2)
new_df
```
