```python
import pandas as pd
df1 = pd.DataFrame({"city": ["new york","chicago","orlando","california"],
    "temperature": [21,14,35,28],})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando","ohio"],
    "humidity": [65,68,75,58],
})
df2
```

```python
merge_df = pd.merge(df1,df2)
merge_df
```

    Merging works like ven diagrams. merging will be done only on common  values

```python
merge_df = pd.merge(df1,df2,how='left')
merge_df
```

```python
merge_df = pd.merge(df1,df2,how='right')
merge_df
```

```python
merge_df = pd.merge(df1,df2,how='outer')
merge_df
```

```python
merge_df = pd.merge(df1,df2,how='inner')
merge_df
```

```python
merge_df = pd.merge(df1,df2,how='union')
merge_df
```

```python
merge_df = pd.merge(df1,df2,how='outer',indicator=True)
merge_df
```

# PIVOTING

Pivoting allows us to summarize & aggregate tabular data

```python
import pandas as pd
import numpy as np

df = pd.read_csv('pivot_weather.csv')
df
```

```python
df.pivot_table(index="city",columns="date")
```

```python
df.pivot_table(index='city',columns='date',values='temperature')
```

```python
df.pivot_table(index='date',columns='city')
```

```python
df.pivot(index='humidity',columns='city')
```

# PIVOT TABLE

```python
import pandas as pd
import numpy as np

df = pd.read_csv("pivot_weather2.csv")
df
```

```python
df.pivot_table(index='city',columns='date')
```

```python
df.pivot(index='city',columns='date') #if DF has duplicate entries we can not reshape it using pivot. we can use 
                                        #pivot_table for duplicate
```

```python
df.pivot_table(index='city',columns='date',margins=True)
```

```python
df.pivot_table(index='city',columns='date',aggfunc=np.sum)
```

```python
df.pivot_table(index='city',columns='date',aggfunc=np.sum,margins=True)
```


# GROUPER

```python
df = pd.read_csv("pivot_weather3.csv")
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city',aggfunc=np.sum)
```
